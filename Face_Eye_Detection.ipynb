Face & Eye Dectetion using HAAR cascade classifier

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Load our image then convert it to grayscale

```python
image = cv2.imread('child.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Input Image')
cv2.imwrite('input_image.jpg', image)
plt.show()
```

We point Open =CV cascasde classifier function to  where our classifier (XML file format) is stored

```python
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')

#OUr classifier returns the ROi of the detected face as a tuple
#It stores the top left coordinate and the bottom right coordiantes
faces = face_classifier.detectMultiScale(gray, 1.3, 5)

#when no faces detected, face_classifier return and empty tupleif faces is():
if faces is ():
  print("No faces found")

#We lterate through our faces array and draw a rectangle
    #over each face in faces
for (x,y,w,h) in faces:
        cv2.rectangle(image, (x,y), (x+w, y+h), (127,0,255), 2)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title('Face Detection')
        cv2.imwrite('face_detection.jpg', image); plt.show()
```
